#### **Relación entre horas de estudio y nota final**

El tiempo dedicado al estudio suele considerarse uno de los factores más relevantes para el rendimiento académico. En este gráfico se analiza la relación entre las **horas de estudio por día** (`study_hours_per_day`) y la **nota final obtenida** (`final_grade`).

Para visualizar esta relación se utiliza un **mapa de densidad (histograma 2D)**, donde cada celda representa la cantidad de estudiantes que se encuentran dentro de un rango específico de horas de estudio y de calificación final.

```python
plt.figure(figsize=(8,5))

h = plt.hist2d(
    df["study_hours_per_day"],
    df["final_grade"],
    bins=(30,30)
)

plt.title("Densidad: horas de estudio vs nota final")
plt.xlabel("Horas de estudio por día")
plt.ylabel("Nota final")
plt.ylim(40,100)

plt.colorbar(h[3], label="Cantidad de estudiantes")

plt.grid(alpha=0.2)

plt.show()
```

**Interpretación**

El mapa de densidad muestra cómo se distribuyen los estudiantes según las horas de estudio y la nota final obtenida. Las zonas más claras indican una mayor concentración de estudiantes dentro de esos rangos de valores.

Se observa que las áreas con mayor concentración tienden a desplazarse hacia notas más altas a medida que aumentan las horas de estudio. Esto sugiere una tendencia positiva moderada entre ambas variables, indicando que dedicar más tiempo al estudio puede asociarse con mejores resultados académicos, aunque también se observa variabilidad en los datos.

#### Relación entre horas de sueño y nota final

El descanso es un factor importante para el bienestar y el desempeño académico de los estudiantes, ya que puede influir en la concentración, la memoria y la capacidad de aprendizaje. Por este motivo, resulta relevante analizar si existe alguna relación entre las horas de sueño y el rendimiento académico.

En el dataset, la variable `sleep_hours` se encuentra registrada con valores decimales, lo que genera una gran cantidad de valores distintos. Si se calculara el promedio de notas para cada valor exacto de horas de sueño, el gráfico resultaría difícil de interpretar debido a la gran cantidad de puntos.

Para facilitar la visualización del patrón general, las horas de sueño se **redondean al valor entero más cercano**, agrupando así a los estudiantes según cantidades aproximadas de horas de descanso. Luego se calcula el **promedio de la nota final para cada grupo de horas de sueño**, lo que permite observar de forma más clara si existe alguna tendencia entre el descanso y el rendimiento académico.

```python
df["sleep_round"] = df["sleep_hours"].round()

sleep_avg = df.groupby("sleep_round")["final_grade"].mean()

plt.figure(figsize=(8,5))

plt.plot(sleep_avg.index, sleep_avg.values, marker="o")

plt.title("Promedio de nota final según horas de sueño")
plt.xlabel("Horas de sueño por día")
plt.ylabel("Nota promedio")

plt.grid(alpha=0.3)

plt.show()
```

**Interpretación**

El gráfico muestra el promedio de la nota final para cada grupo de horas de sueño. Cada punto representa la calificación promedio obtenida por los estudiantes que duermen aproximadamente esa cantidad de horas por día.

A partir de la visualización se observa que el promedio de las notas tiende a aumentar levemente a medida que aumentan las horas de sueño. Aunque la variación no es muy marcada, se puede identificar una ligera tendencia positiva que sugiere que dormir más horas podría asociarse con mejores resultados académicos.

Sin embargo, las diferencias entre los promedios no son muy pronunciadas, lo que indica que las horas de sueño por sí solas no explican completamente el rendimiento académico de los estudiantes.

Sin embargo, se identifican algunas asociaciones moderadas entre otras variables del dataset. Por ejemplo, se observa una relación positiva entre **horas de estudio (`study_hours_per_day`) y productividad (`productivity_score`)**, lo que sugiere que los estudiantes que dedican más tiempo al estudio tienden a presentar mayores niveles de productividad.
